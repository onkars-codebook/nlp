# 1. using *NLTK*

```python
import re
import nltk

# Download required NLTK resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python

#input
text = "My name is Sanket Mahajan , currently persuing Bachlor of technology in AIDS. "
print(text)
```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt_tab')
```

```python
words = word_tokenize(text)
print(words)
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
print(sentences)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_words]
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print(lemmatized_words)
```

# 2. Reglar Expression

```python
# Original text
text = "I have interest in AI! Machine learning is cool, right? Contact: seses123444@yahoo.com"

print("Original Text:")
print(text)
print("-" * 50)
```

```python
# 1. Lowercasing
text = text.lower()
print("After Lowercasing:")
print(text)
print("-" * 50)
```

```python
# 2. Remove email IDs using regex
text = re.sub(r'\S+@\S+', '', text)
print("After Removing Emails:")
print(text)
print("-" * 50)
```

```python
# 3. Remove punctuation using regex
text = re.sub(r'[^\w\s]', '', text)
print("After Removing Punctuation:")
print(text)
print("-" * 50)
```

```python
# 4. Remove numbers using regex
text = re.sub(r'\d+', '', text)
print("After Removing Numbers:")
print(text)
print("-" * 50)
```

```python
# 5. Tokenization using regex
tokens = re.findall(r'\b\w+\b', text)
print("After Tokenization:")
print(tokens)
print("-" * 50)
```

```python
# 6. Stop word removal
stop_words = {'i', 'am', 'is', 'it'}
filtered_tokens = [word for word in tokens if word not in stop_words]

print("After Stop Word Removal:")
print(filtered_tokens)
print("-" * 50)
```

```python
# 7. Stemming using simple regex rules
def stem(word):
    return re.sub(r'(ing|ed|s)$', '', word)

stemmed_words = [stem(word) for word in filtered_tokens]

print("After Stemming:")
print(stemmed_words)
print("-" * 50)

```

```python
# 8. Lemmatization (rule-based approximation)
lemma_rules = {
    'learning': 'learn',
    'amazing': 'amaze',
    'emails': 'email'
}

lemmatized_words = [lemma_rules.get(word, word) for word in filtered_tokens]

print("After Lemmatization:")
print(lemmatized_words)
print("-" * 50)
```
