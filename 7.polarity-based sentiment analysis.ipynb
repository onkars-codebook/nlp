```python
# Perform single-word, multi-word based and polarity-based sentiment analysis
# Polarity means direction of feeling: positive (+), negative (-), or neutral (0).
# We score words/phrases, then use the final score direction as sentiment.
# Step 1: Collect Text Data

text = input("Enter a review/sentence: ")

# Example:
# This product is very good and the design is nice
```

```python
# Step 2: Preprocessing

import re

# convert to lowercase
text = text.lower()

# remove punctuation
text = re.sub(r'[^\w\s]', '', text)

# tokenize
words = text.split()

print("Tokenized Words:", words)
```

```python
# Step 3: Single-word Sentiment Analysis
# This is a lexicon-based method: we keep simple lists of positive/negative words.

positive_words = {
    "good", "great", "excellent", "nice", "amazing", "happy",
    "love", "best", "wonderful", "awesome"
}

negative_words = {
    "bad", "poor", "terrible", "awful", "worst", "sad",
    "hate", "boring", "disappointing"
}

positive_score = 0
negative_score = 0

for word in words:
    if word in positive_words:
        positive_score += 1
    elif word in negative_words:
        negative_score += 1

print("Single Word Analysis")
print("Positive Score:", positive_score)
print("Negative Score:", negative_score)
```

```python
# Step 4: Multi-word Sentiment Analysis
# Some emotions appear in 2-word phrases, so we check pairs of neighboring words too.

multi_word_phrases = {
    "very good": 2,
    "very nice": 2,
    "not good": -2,
    "not bad": 1,
    "very bad": -2
}

multi_score = 0

for i in range(len(words)-1):
    phrase = words[i] + " " + words[i+1]

    if phrase in multi_word_phrases:
        multi_score += multi_word_phrases[phrase]
        print("Detected phrase:", phrase)

print("Multi-word Score:", multi_score)
```

```python
# Step 5: Polarity Calculation
# It is called polarity-based because we care about the sign of total score.
# Positive sign => Positive sentiment, negative sign => Negative sentiment, zero => Neutral.

total_score = positive_score - negative_score + multi_score

print("Total Sentiment Score:", total_score)

if total_score > 0:
    sentiment = "Positive"
elif total_score < 0:
    sentiment = "Negative"
else:
    sentiment = "Neutral"

print("Sentiment:", sentiment)
```

```python
# Step 6: Result Interpretation

print("\nFinal Analysis")
print("-----------------------")
print("Sentence:", text)
print("Positive Words:", positive_score)
print("Negative Words:", negative_score)
print("Phrase Score:", multi_score)
print("Final Sentiment:", sentiment)
```
