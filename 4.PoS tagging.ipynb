```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import RegexpTagger
import spacy

# Sample text (you can modify this)
text = """Artificial Intelligence is transforming the world rapidly.Machine learning models are becoming more powerful and efficient."""
```

```python
import nltk
nltk.download('punkt_tab')
# Tokenize text using NLTK
tokens = word_tokenize(text)

print("Tokens:")
print(tokens)
```

```python
# Define RegEx patterns
patterns = [
    (r'.*ing$', 'VBG'),      # gerunds
    (r'.*ed$', 'VBD'),       # past tense verbs
    (r'.*es$', 'VBZ'),       # 3rd person singular verbs
    (r'.*ould$', 'MD'),      # modals
    (r'.*\'s$', 'NN$'),      # possessive nouns
    (r'.*s$', 'NNS'),        # plural nouns
    (r'^-?[0-9]+(\.[0-9]+)?$', 'CD'),  # numbers
    (r'.*', 'NN')            # default noun
]

regexp_tagger = RegexpTagger(patterns)

regex_tagged = regexp_tagger.tag(tokens)

print("POS Tagging using Regular Expressions:")
print(regex_tagged)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
nltk_tagged = nltk.pos_tag(tokens)

print("POS Tagging using NLTK Inbuilt Tagger:")
print(nltk_tagged)
```

```python
# Perform PoS tagging using regular expressions and inbuilt PoS taggers
# Load spaCy model
nlp = spacy.load("en_core_web_sm")

doc = nlp(text)

print("POS Tagging using spaCy:")
for token in doc:
    print(token.text, "->", token.pos_, )
```
