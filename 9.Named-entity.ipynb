Recognize Named Entities from a given paragraph.

```python
import spacy
from spacy import displacy
```

```python
import spacy
```

```python
# Load the model
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Virat Kohli scored a century for India in Mumbai on 5 March 2023."
```

```python
doc = nlp(text)
```

```python
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
from spacy import displacy

displacy.render(doc, style="ent", jupyter=True)
```
